```python
#Q3
mq3 = np.trunc(np.dot(np.random.rand(5, 5), 100))

def normalize_with_list(matrix):
    return np.matrix(list(map(lambda x: (x - x.mean()) / x.std(), mq3.T))).T

def normalize_without_list(matrix):
    result = np.zeros(matrix.shape)
    for i in range(matrix.shape[1]):
        c = matrix[:, i]
        result[:, i] = (c - c.mean()) / c.std()
    return result
print(mq3, normalize_with_list(mq3), normalize_without_list(mq3), sep='\n\n')
```
